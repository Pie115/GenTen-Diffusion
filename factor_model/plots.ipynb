```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rank_axis = [5, 10, 25, 50, 100, 150, 200, 250]
FID_axis = [53.59, 53.89, 37.14, 21.12, 16.01, 19.04, 18.71, 25.66]
```

```python
plt.figure(figsize=(6, 4))
plt.plot(rank_axis, FID_axis, color='black', linewidth=2)
plt.xlabel('Rank', fontsize=12)
plt.ylabel('FID Score', fontsize=12)
plt.title('Rank vs FID', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 65)

plt.text(0.5, -0.18, '(b) Diffusion', ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)

plt.tight_layout()

plt.savefig("FID_Rank_Diffusion")
```

```python
parameter_axis = [((51 + 51 + 25)*r)/(51*51*25)*100 for r in [5, 10, 25, 50, 100, 150, 200, 250]]

FID_axis = [48.12, 44.12, 41.38, 18.99, 18.37, 18.30, 18.19, 16.52]

plt.figure(figsize=(10, 6))
plt.plot(parameter_axis, FID_axis, linewidth = 5, color = 'black')
plt.ylim(bottom = 0)


axis_font_size = 35
plt.xlabel('Percent of Full Output Parameters', fontsize = axis_font_size)
plt.ylabel('FID Score', fontsize = axis_font_size)
plt.title('Parameters vs FID', fontsize = 50)

tick_font_size = 20
xticks = list(range(0, 50+1, 10))
plt.xticks(xticks, labels = [f'{xt}%' for xt in xticks], fontsize = tick_font_size)
plt.yticks(list(range(0, 50+1, 5)), fontsize=tick_font_size)

plt.savefig("Diffusion_parameters_vs_FID", bbox_inches='tight', dpi=300)
```
