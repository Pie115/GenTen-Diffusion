```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rank_axis = [5, 10, 25, 50, 100, 150, 200, 250]
FID_axis = [53.59, 53.89, 37.14, 21.12, 16.01, 19.04, 18.71, 25.66]
```

```python
plt.figure(figsize=(6, 4))
plt.plot(rank_axis, FID_axis, color='black', linewidth=2)
plt.xlabel('Rank', fontsize=12)
plt.ylabel('FID Score', fontsize=12)
plt.title('Rank vs FID', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0, 65)

plt.text(0.5, -0.18, '(b) Diffusion', ha='center', va='center', transform=plt.gca().transAxes, fontsize=12)

plt.tight_layout()

plt.savefig("FID_Rank_Diffusion")
```
